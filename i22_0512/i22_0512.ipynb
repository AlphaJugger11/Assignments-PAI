Let me provide you with an example of how to apply a simple grayscale filter on an image.

When transforming an RGB image into grayscale, we calculate the average of the RGB values for each pixel. The array holding the grayscale image has two dimensions, unlike the RGB image which has three. This distinction arises because in a grayscale image, the three RGB values are substituted with a single average value derived from the Red, Green, and Blue values of the pixel.

```python
grayscale_image = []
# Iterating Over Rows of Pixels in the Image
for i in range(len(img_array)):
    temp = []
    # Getting The Values from Each Pixel from The Row
    for j in range(len(img_array[i])):
        # Get the Average Value of RGB Values from the Pixel and append it to a temporary list
        avg = int(sum(img_array[i][j])/3)
        temp.append(avg)
    # append the row of average pixel values to a new array
    grayscale_image.append(temp)
```

In image processing, you often need to extract special features from images. One such feature can be the object boundaries/edges in the images. One effective technique to accomplish this task involves the utilization of predefined kernels. Kernels, simply put, are matrices which are multiplied to the image array.

```python
# X Kernel

Gx = [
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
]
```

```python
# Y Kernel

Gy = [
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
]
```

#### Pseudo Code

Create an empty result image (list)

for each pixel row in the grayscale image:
    for each pixel column in the grayscale image:
        Initialize vertical gradient and horizontal gradient as 0
        
        for each kernel row:
            for each kernel column:
                Calculate the corresponding pixel coordinates in the image
                Multiply the kernel value with the pixel value at the coordinates
                Add the result to the vertical gradient and horizontal gradient
                
        Calculate the gradient magnitude as sqrt(vertical_gradient^2 + horizontal_gradient^2)
        Set the pixel value in the result image as the gradient magnitude
        

In simpler terms, for each pixel in the grayscale image form a 3x3 grid of pixels by taking pixels from neighbouring rows and columns of the pixel. Next, multiply and add the values of the pixels from this grid to its corresponding element in the kernels, once for the horizontal direction (Gx) and once for the vertical direction (Gy). Once you've finished these calculations, you'll have two numbers. Square both of these numbers, add them together, and then find the square root of the total. If this final number is larger than 255, set it to 255. Take this new number and replace the original pixel value with it in the result image (do not change the original grayscale image).
