```python
 """Reading original excel files for matching"""
    
for_color_only = pd.read_excel('TimeTable, FSC, Fall-2023.xlsx', 'Monday')
for_color_only.fillna('Break',inplace=True)
for_color_only.reset_index(drop=True,inplace=True)
for_color_only.index = range(1, len(for_color_only)+1)


tue_orig = pd.read_excel('TimeTable, FSC, Fall-2023.xlsx','Tuesday')
tue_orig.fillna('Break',inplace=True)
tue_orig.reset_index(drop=True,inplace=True)
tue_orig.index = range(1,len(tue_orig)+1)



wed_orig = pd.read_excel('TimeTable, FSC, Fall-2023.xlsx','Wednesday')
wed_orig.fillna('Break',inplace=True)
wed_orig.reset_index(drop=True,inplace=True)
wed_orig.index = range(1,len(wed_orig)+1)


thu_orig = pd.read_excel('TimeTable, FSC, Fall-2023.xlsx','Thursday')
thu_orig.fillna('Break',inplace=True)
thu_orig.reset_index(drop=True,inplace=True)
thu_orig.index = range(1,len(thu_orig)+1)

thu_orig
```

```python
###DO not change this one ##########

import openpyxl as opx
batch = "BS AI (2022)"
batch = input("Which batch are you from enter your batch and department in correct format e.g BS AI (2022)  ")
Timetable = opx.load_workbook('TimeTable, FSC, Fall-2023.xlsx')
MONday = Timetable['Monday']
TUESday = Timetable['Tuesday']
WEDnesday = Timetable['Wednesday']
THUsday = Timetable['Thursday']
   
##  THIS Part is for getting the color of batch that has been entered by user

rowforcolor = 0
colforcolor = 0
for rows, indexes in for_color_only.iterrows():
    for i in range(len(indexes)):
        if batch in indexes[i]:
            rowforcolor,colforcolor = rows,i
#             print(rowforcolor, "  " , colforcolor)
# print(rowforcolor, "  " , colforcolor)

cel_to_check = MONday.cell(row=rowforcolor+1,column=colforcolor+1)
RGB_of_batch= ''
if cel_to_check.fill.start_color.index is not None:
    RGB_of_batch = cel_to_check.fill.start_color.rgb

l_for_batchtt_mon = []
l_for_batchtt_tue = []
l_for_batchtt_wed = []
l_for_batchtt_thu = []

## the implementating
###                All of the implementation is for color matching 
for row in MONday.iter_rows(min_row=1,max_row=MONday.max_row, min_col=1,max_col=MONday.max_column):
    for dcell in row:
        if dcell.fill.start_color.index is not None:
            rgb_of_curr_cell = dcell.fill.start_color.rgb
            if rgb_of_curr_cell == RGB_of_batch:
                l_for_batchtt_mon.append([dcell.row-2,dcell.column-1])
                


for row in TUESday.iter_rows(min_row=1,max_row=TUESday.max_row, min_col=1,max_col=TUESday.max_column):
    for dcell in row:
        if dcell.fill.start_color.index is not None:
            rgb_of_curr_cell = dcell.fill.start_color.rgb
            if rgb_of_curr_cell == RGB_of_batch:
                l_for_batchtt_tue.append([dcell.row-2,dcell.column-1])


for row in WEDnesday.iter_rows(min_row=1,max_row=WEDnesday.max_row, min_col=1,max_col=WEDnesday.max_column):
    for dcell in row:
        if dcell.fill.start_color.index is not None:
            rgb_of_curr_cell = dcell.fill.start_color.rgb
            if rgb_of_curr_cell == RGB_of_batch:
                l_for_batchtt_wed.append([dcell.row-2,dcell.column-1])


for row in THUsday.iter_rows(min_row=1,max_row=THUsday.max_row, min_col=1,max_col=THUsday.max_column):
    for dcell in row:
        if dcell.fill.start_color.index is not None:
            rgb_of_curr_cell = dcell.fill.start_color.rgb
            if rgb_of_curr_cell == RGB_of_batch:
                l_for_batchtt_thu.append([dcell.row-2,dcell.column-1])
```

```python
tt_for_mon = []
tt_for_tue = []
tt_for_wed = []
tt_for_thu = []

'''what this code does is that it takes the those indexes at stores data at those indexes in the lists above'''

for i in l_for_batchtt_mon:
    tt_for_mon.append(for_color_only.iloc[i[0],i[1]])


for i in l_for_batchtt_tue:
    tt_for_tue.append(tue_orig.iloc[i[0],i[1]])
        

for i in l_for_batchtt_wed:
    tt_for_wed.append(wed_orig.iloc[i[0],i[1]])

for i in l_for_batchtt_thu:
    tt_for_thu.append(thu_orig.iloc[i[0],i[1]])


# ok so what we can do is that we got all values of the same batch so we can complete it
#tt_for_tue
```

```python
dfmon = pd.read_excel("TimeTable, FSC, Fall-2023.xlsx", 'Monday'  )
dftues = pd.read_excel("TimeTable, FSC, Fall-2023.xlsx", 'Tuesday')
dfwed = pd.read_excel("TimeTable, FSC, Fall-2023.xlsx", 'Wednesday')
dfthu = pd.read_excel("TimeTable, FSC, Fall-2023.xlsx", 'Thursday')

'''making an empty data frame so I can read fill it with data of respective batch'''

mon = pd.DataFrame(index=range(41), columns=dfmon.loc[3])
tue = pd.DataFrame(index=range(42),columns=dftues.loc[3])
wed = pd.DataFrame(index=range(44),columns=dfwed.loc[3])
thu = pd.DataFrame(index=range(59),columns=dfthu.loc[3])

'''Code below changes the indexes column to Rooms from original dataframe'''

for i in range(len(mon)):
    mon['Room'][i] = dfmon['Monday'][i]
for i in range(len(tue)):
    tue['Room'][i] = dftues['Tuesday'][i]
for i in range(len(wed)):
    wed['Room'][i] = dfwed['Wednesday'][i]
for i in range(len(thu)):
    thu['Room'][i] = dfthu['Thursday'][i]

'''The code below puts values of classes by index found above in l_for_batchtt_DAY and puts the class room at that index '''
for i in range(len(l_for_batchtt_mon)):
        mon.iat[l_for_batchtt_mon[i][0],l_for_batchtt_mon[i][1]] = tt_for_mon[i]

        
for i in range(len(l_for_batchtt_tue)):
        tue.iat[l_for_batchtt_tue[i][0],l_for_batchtt_tue[i][1]] = tt_for_tue[i]

for i in range(len(l_for_batchtt_wed)):
        wed.iat[l_for_batchtt_wed[i][0],l_for_batchtt_wed[i][1]] = tt_for_wed[i]
        
for i in range(len(l_for_batchtt_thu)):
        thu.iat[l_for_batchtt_thu[i][0],l_for_batchtt_thu[i][1]] = tt_for_thu[i]
        
'''Finally filling the data frames nan value with - so easier to visualize '''
mon.fillna("-",inplace=True)
tue.fillna("-",inplace=True)
wed.fillna("-",inplace=True)
thu.fillna("-",inplace=True)

thu
```

```python
'''Writing data into excel file'''
with pd.ExcelWriter('SE(2022).xlsx') as wr:
    mon.to_excel(wr,sheet_name="Monday",index=False)
    tue.to_excel(wr,sheet_name='Tuesday',index=False)
    wed.to_excel(wr,sheet_name='Weednesday',index=False)
    thu.to_excel(wr,sheet_name='Thursday',index=False)
```
